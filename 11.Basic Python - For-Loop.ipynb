```python
def compute_interest (money , period ):
    #period: số kỳ nhận lãi trong năm (12 nếu theo tháng, 365 nếu theo ngày, 730 nếu 2 lần/ngày,...)
    amount = money
    for _ in range(period):
        amount *= (1 + 1 / period)
    return amount
```

```python
print(compute_interest(1, 12))
print(compute_interest(1, 365))
print(compute_interest(1, 730))
```

# Bài tập 2

image.png

```python
def is_prime_eratosthenes(n):
    if n < 2:
        return False

    prime = []
    for i in range(0, n + 1):
        prime.append(True)
    prime[0] = False
    prime[1] = False

    for i in range(2, int(n ** 0.5) + 1):
        if prime[i] == True:
            for j in range(i * i, n + 1, i):
                prime[j] = False
    return prime[n]
```

```python
print(is_prime_eratosthenes(7))
print(is_prime_eratosthenes(10))
print(is_prime_eratosthenes(2))
print(is_prime_eratosthenes(1))
print(is_prime_eratosthenes(9))
```

# Bài tập 3

image.png

```python
def fizz_buzz(n):
    for i in range(1, n+1):
        if i % 3 == 0 and i % 5 == 0:
            print("Fizz-Buzz", end=", ")
        elif i % 3 == 0:
            print("Fizz", end=", ")
        elif i % 5 == 0:
            print("Buzz", end=", ")
        else:
            print(i, end=", ")
```

```python
fizz_buzz(3)
fizz_buzz(5)
fizz_buzz(15)
```

# Bài tập 4

image.png

```python
def factorial(n):
    result = 1
    for data in range(1, n+1):
        result *= data
    return result
```

```python
factorial(5)
```

```python
def calculate_total(n):
    odd = 0
    even = 0
    for data in range(1, n + 1):
        if data % 2 == 0:
            even += data
        else:
            odd += data
    print(f"odd={odd}, event={even}")
```

```python
calculate_total(5)
```
