# 1.Chuẩn hóa dữ liệu (Normalization)

image.png

```python
def normalize ( lst_data ):
  min_x = min(lst_data)
  max_x = max(lst_data)

  result = []
  for data in lst_data:
    result.append((data - min_x) / (max_x - min_x))

  return result
```

```python
test_cases = [
    [11 , 18, 24, 30, 36] ,
    [50 , 100 , 150 , 200 , 250] ,
    [3, 5, 7, 9, 11]
]
```

```python
for data in test_cases:
  print(normalize(data))
```

# 2. Trung bình động (Moving Average)

image.png
image.png

```python
def moving_average (lst , k):
    result = []
    for idx, data in enumerate (lst):
        if idx - k + 1 < 0 or idx - k + 1 > len(lst) - 1:
            continue
        result.append(sum(lst[idx - k + 1 : idx + 1]) / k)

    return result
```

```python
test_cases = [
    ([1 , 2, 3, 4, 5, 6, 7, 8, 9], 3) ,
    ([10 , 20, 30, 40, 50, 60, 70] , 4) ,
    ([5 , 10, 15, 20, 25] , 2)
]
```

```python
for i, (test , k) in enumerate(test_cases):
    result = moving_average(test , k)
    print (f" Test {i + 1}: {result}")
```

# 3. Tích vô hướng của hai vector

image.png

```python
def dot_product (v1 , v2):
    result = sum(i * j for i, j in zip(v1 , v2))
    return result
```

```python
test_cases = [
    ([1 , 2, 3], [4, 5, 6]) ,
    ([2 , 4, 6], [1, 3, 5]) ,
    ([0 , 1, 2], [3, 4, 5])
]
```

```python
for i, (v1 , v2) in enumerate(test_cases):
    result = dot_product(v1 , v2)
    print (f" Test {i + 1}: {result}")
```

# 4. Mô phỏng Perceptron

image.png
image.png

```python
def perceptron_relu(weights , inputs , bias):
    result = sum(x * y for x, y in zip(weights, inputs)) + bias
    return result
```

```python
test_cases = [
    ([0.5, -0.6, 0.8], [1.0, 0.0, 1.0] , -0.3) ,
    ([0.2, 0.5, -0.4], [1.0, 2.0, -1.0] , 0.1) ,
    ([1.0, -1.0, 0.5], [0.5, 1.0, -0.5] , -0.2)
]
```

```python
for i, (weights, inputs, bias) in enumerate(test_cases):
    result = perceptron_relu(weights, inputs , bias)
    print (f" Test {i + 1}: {result}")
```
