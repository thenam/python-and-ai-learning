```python
my_tuple1 = (2,3)
my_tuple2 = (3,6)
```

```python
# Nếu vector a = (a1, a2) và vector b = (b1, b2), thì tổng của chúng, vector c = a + b, sẽ là c = (a1 + b1, a2 + b2).
total_tuple = (my_tuple1[0] + my_tuple2[0], my_tuple1[1] + my_tuple2[1])
# Nếu vector a = (x1, y1) và vector b = (x2, y2), thì tích của chúng là a · b = (x1 * x2, y1 * y2).
# Nếu vector a = (x1, y1) và vector b = (x2, y2), thì tích vô hướng của chúng là a · b = x1x2 + y1y2.
muti_tuple = (my_tuple1[0] * my_tuple2[0], my_tuple1[1] * my_tuple2[1])
```

```python
print(total_tuple)
print(muti_tuple)
```

```python
import math

distance_tuple = math.sqrt((my_tuple1[0] - my_tuple2[0]) ** 2 + (my_tuple1[1] - my_tuple2[1]) ** 2)
print(distance_tuple)
```
