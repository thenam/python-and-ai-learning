# Bài tập: Bag of Words

image.png

```python
corpus = ["Tôi thích môn Toán", "Tôi thích AI", "Tôi thích âm nhạc"]
```

```python
words = []
for doc in corpus:
  print(doc)
  words = words + doc.split()

print(words)
```

```python
id_words = []
for word in words:
  if word not in id_words:
    id_words.append(word)

print(id_words)
```

```python
input_str = "Tôi thích AI thích Toán"
```

```python
input_words = input_str.split()

print(input_words)
```

```python
bow = []
for idw in id_words:
  count = 0
  for word in input_words:
    if idw == word:
      count = count + 1
  bow.append(count)

print(bow)
```
