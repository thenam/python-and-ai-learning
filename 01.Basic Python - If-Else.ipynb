# Tính lịch Can Chi - làm quen với câu lệnh If/Else

image.png

```python
def calculate_can_chi_calendar(year):
  # Cách 1: if else 22 lần
  # Cách 2: dùng Tuple
  result = ''
  lst_can = ('Canh','Tân','Nhâm','Quý','Giáp','Ất','Bính','Đinh','Mậu','Kỷ')
  lst_chi = ('Thân','Dậu','Tuất','Hợi','Tý','Sửu','Dần','Mẹo','Thìn','Tỵ','Ngọ','Mùi')
  can = year % 10
  chi = year % 12
  result = lst_can[can] + ' ' + lst_chi[chi]
  return result
```

```python
assert calculate_can_chi_calendar(2025) == 'Ất Tỵ', 'Tính sai'
assert calculate_can_chi_calendar(2024) == 'Giáp Thìn', 'Tính sai'
assert calculate_can_chi_calendar(1997) == 'Đinh Sửu', 'Tính sai'

print(calculate_can_chi_calendar(2025))
print(calculate_can_chi_calendar(2024))
print(calculate_can_chi_calendar(1997))
```

```python
import math

def sigmoid(x):
  return 1 / (1 + math.exp(-x))

print(sigmoid(0))
print(sigmoid(2))
```

```python
import math

def tank(x):
  return (math.exp(x) - math.exp(-x))/(math.exp(x) + math.exp(-x))

print(tank(0))
print(tank(2))
print(tank(150))
```

```python
import math

def elu(x, alpha = 0.01):
  if x > 0:
    return x
  else:
    return alpha * (math.exp(x) - 1)

print(elu(3))
print(elu(-1))
print(elu(0))
```
