```python
with open("input-data.txt", "r") as f:
    data = f.read()
    print(data)
```

```python
data = data.replace("\n", " ")
data
```

```python
data = data.lower()
data
```

```python
lst_data = data.split()
lst_bow = []
lst_tokenized = []
for word in lst_data:
    if word not in lst_bow:
        lst_bow.append(word)
        lst_tokenized.append(1)
    else:
        idx = lst_bow.index(word)
        lst_tokenized[idx] += 1
```

```python
print(lst_bow)
print(lst_tokenized)
```

```python
max_freq = max(lst_tokenized)
max_freq
min_freq = min(lst_tokenized)
min_freq
```

```python
idx_max_freq = lst_tokenized.index(max_freq)
lst_bow[idx_max_freq]
```

```python
idx_min_freq = lst_tokenized.index(min_freq)
lst_bow[idx_min_freq]
```
