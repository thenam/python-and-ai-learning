```python
ma = [[1,2,3],[4,5,6],[7,8,9]]
mb = [[2,4,6],[1,3,5],[1,0,1]]
```

```python
tong = []
for i in range(len(ma)):
  sum = []
  for j in range(len(ma[i])):
    sum.append(ma[i][j] + mb[i][j])
  tong.append(sum)

print(tong)
```

```python
hieu = []
for i in range(len(ma)):
  sub = []
  for j in range(len(ma[i])):
    sub.append(ma[i][j] - mb[i][j])
  hieu.append(sub)

print(hieu)
```

```python
tich = []
for i in range(len(ma)):
  mul = []
  for j in range(len(mb)):
    sum = 0
    for k in range(len(mb[j])):
      sum = sum + ma[i][k] * mb[k][j]
    mul.append(sum)

  tich.append(mul)

print(tich)
```

```python
mc = [[1,2], [3,4], [4,5]]
rm = [[1,1,1], [1,1,1]]

tich = []
for i in range(len(mc)):
  mul = []
  for j in range(len(rm)):
    sum = 0
    for k in range(len(rm)):
      sum = sum + mc[i][k] * rm[k][j]
    mul.append(sum)

  tich.append(mul)

print(tich)
```
