```python
lst_data = [
  [1, 2, 3],
  [4, 5, 6],
  [7, 8, 9],
]
```

```python
lst_sub_data = []
for data in lst_data:
  lst_sub = []
  for idx in range(len(data)):
    if idx == 0 or idx == 2:
      lst_sub.append(data[idx])
  lst_sub_data.append(lst_sub)

print(lst_sub_data)
```
