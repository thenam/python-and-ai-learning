```python
lst_data = [1, 1.1, None, 1.4, None, 1.5, None, 2.0].
lst_data = [None, None, None, 1.4, None, 1.5, None, 2.0]
```

```python
for idx in range(0, len(lst_data)):
  if lst_data[idx] == None:
    if idx - 1 >= 0 and lst_data[idx - 1] != None:
      lst_data[idx] = lst_data[idx - 1]
    elif idx + 1 < len(lst_data) and lst_data[idx + 1] != None:
      lst_data[idx] = lst_data[idx + 1]

print(lst_data)
```
