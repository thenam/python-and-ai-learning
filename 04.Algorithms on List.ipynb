```python
lst_data = [1, 1.1, None, 1.4, None, 1.5, None, 2.0]
```

```python
lst_none = []

for idx in range(len(lst_data)):
  if lst_data[idx] == None:
    lst_none.append(idx)
```

```python
if len(lst_none) > 0:
  print(f"Vị trí None đầu tiên: {lst_none[0]}")

print(f"Danh sách vị trí có giá trị None: {lst_none}")
```
