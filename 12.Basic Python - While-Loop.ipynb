```python
def find_divisible_number(a):

    result = 100
    while(True):
        result += 1
        if result % a == 0:
            break

    return result
```

```python
assert find_divisible_number(5) == 105
```

```python
assert find_divisible_number(17) == 102
```

```python
def find_squared_root(a):

    EPSILON = 0.001
    n = 1
    n1 = 0
    while(True):
        n1 = (n + a / n) / 2
        if abs(n1 - n) < EPSILON:
            break

        n = n1

    return n1
```

```python
find_squared_root(2)
```

```python
find_squared_root(3)
```

```python
import random

def random_number_with_condition(total):
    # Set a random value that is the same between devices
    random.seed (0)

    count = 1
    while(True):
        a = random.randint(1, 20)
        b = random.randint(1, 20)
        if a + b != total:
            count += 1
        else:
            break

    return count
```

```python
random_number_with_condition(40)
```

```python
random_number_with_condition(20)
```
